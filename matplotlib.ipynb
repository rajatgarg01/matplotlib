```python
#making some data
nut_butter_price={"almond butter":10,"peanut butter":8,"cashew butter":12}
fig,ax=plt.subplots();
ax.bar(nut_butter_price.keys(),nut_butter_price.values())
ax.set(title="dan's nut butter store",
       ylabel="price$")
fig.savefig("sample3")
```

```python
#horizontal bar
fig,ax=plt.subplots();
ax.barh(nut_butter_price.keys(),nut_butter_price.values())
ax.set(title="dan's nut butter store",
       xlabel="price$")
fig.savefig("sample4")
```

```python
# making some data
x=np.random.random(10)
fig,ax=plt.subplots()
ax.hist(x);
fig.savefig("sample5")
```
